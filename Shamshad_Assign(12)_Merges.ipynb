```python
import pandas as pd

# Sample DataFrames

df1 = pd.DataFrame({
    'key': ['A', 'B', 'C', 'D'],
    'value1': [1, 2, 3, 4]
})

df2 = pd.DataFrame({
    'key': ['B', 'C', 'D', 'E'],
    'value2': [5, 6, 7, 8]
})

# Inner join

inner = pd.merge(df1, df2, how='inner', on='key')
inner
```

```python
# Left join

left = pd.merge(df1, df2, how='left', on='key')
left
```

```python
# Right join

right = pd.merge(df1, df2, how='right', on='key')
right
```

```python
# Outer join

outer = pd.merge(df1, df2, how='outer', on='key')
outer
```

```python
# Cross join

cross = pd.merge(df1, df2, how='cross')
cross
```
